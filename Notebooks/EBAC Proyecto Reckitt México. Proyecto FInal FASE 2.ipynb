```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar los datos

reckitt_df = pd.read_csv('C:/Users/Isaac/Desktop/IHD/EBAC DT/CIENCIA DE DATOS/PROYECTO RECKITT/ReckittMex.csv')
reckitt_df.head(10)
```

```python
reckitt_df.info()
```

### <code> Visualizaciones para Comprender la Distribución de los Datos<code/>

```python
# Tendencias Temporales (DATE vs Ventas)
reckitt_df['DATE'] = pd.to_datetime(reckitt_df['DATE'])  # Convertir DATE a formato datetime

plt.figure(figsize=(12, 6))
sns.lineplot(data=reckitt_df, x='DATE', y='TOTAL_UNIT_SALES', label='Unidades Vendidas', color='blue')
sns.lineplot(data=reckitt_df, x='DATE', y='TOTAL_VALUE_SALES', label='Valor Total Ventas', color='green')
plt.title('Tendencia Temporal de Ventas')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.legend()
plt.show()
```

### <code><a>Insights<a/><code/> 
    
    * En el gráfico, se observan picos estacionales en ciertas fechas (quizás relacionadas con promociones, festividades o ciclos comerciales).

```python
# Comparación entre Regiones (REGION vs Ventas)
plt.figure(figsize=(7, 5))
sns.boxplot(data=reckitt_df, x='REGION', y='TOTAL_UNIT_SALES', palette='Set3')
plt.title('Variación de Ventas Unitarias por Región')
plt.xlabel('Región')
plt.ylabel('Unidades Vendidas')
plt.xticks(rotation=45)
plt.show()
```

### <code><a>Insights<a/><code/> 
    *  Hay una región ("CANNING MEXICO") con un número de ventas unitarias significativamente mayor en comparación con otras regiones, lo que podría indicar:
    * Un mercado más grande.
    * Eficiencias en distribución o promociones agresivas.
    * La mayoría de las otras regiones tienen ventas más uniformes y con menos outliers, lo que puede reflejar estabilidad pero también menor potencial de crecimiento.

```python
# Ventas por Categoría y Segmento
plt.figure(figsize=(12, 6))
sns.barplot(data=reckitt_df, x='CATEGORY_DESCRIPTION', y='TOTAL_VALUE_SALES', hue='SEGMENT', ci=None, palette='muted')
plt.title('Ventas por Categoría y Segmento')
plt.xlabel('Categoría')
plt.ylabel('Valor Total de Ventas')
plt.xticks(rotation=45)
plt.legend(title='Segmento', bbox_to_anchor=(1, 1))
plt.show()
```

### <code><a>Insights<a/><code/> 
    * Las categorías como "LAVAVAJILLAS" y "AIR CARE" lideran en términos de valor total de ventas, especialmente en segmentos como "BLEACH" y "LIQUID & GEL".
    * Algunos segmentos como "SANITIZER" y "PRETREAT" tienen valores más bajos, lo que sugiere:
    Menor preferencia del consumidor o posibles problemas de comercialización.


```python
# Relación entre Ventas Unitarias y Ventas Totales
plt.figure(figsize=(7, 5))
sns.scatterplot(data=reckitt_df, x='TOTAL_UNIT_SALES', y='TOTAL_VALUE_SALES', hue='REGION', palette='coolwarm', alpha=0.7)
plt.title('Relación entre TOTAL_UNIT_SALES y TOTAL_VALUE_SALES')
plt.xlabel('Unidades Vendidas')
plt.ylabel('Valor Total de Ventas')
plt.legend(title='Región', bbox_to_anchor=(1, 1))
plt.show()
```

### <code><a>Insights<a/><code/> 
     * La correlación entre las ventas unitarias y el valor total de las ventas puede indicar una relación lineal positiva, donde productos con alto volumen vendido generan un mayor ingreso.
     * Es posible que haya productos específicos con valores atípicos donde el precio por unidad sea extremadamente alto o bajo, afectando la linealidad.

```python
#  Distribución por Mes y Año
monthly_sales = reckitt_df.groupby(['YEAR', 'MONTH'])['TOTAL_VALUE_SALES'].sum().reset_index()

# Pivot para un formato adecuado
pivot_sales = monthly_sales.pivot(index='MONTH', columns='YEAR', values='TOTAL_VALUE_SALES')

pivot_sales.plot(kind='bar', stacked=True, figsize=(12, 6), cmap='tab20')
plt.title('Ventas Totales Mensuales por Año')
plt.xlabel('Mes')
plt.ylabel('Ventas Totales')
plt.xticks(range(12), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.legend(title='Año', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### <code><a>Insights<a/><code/> 
     * Las ventas parecen tener un patrón mensual/estacional significativo, con ciertos meses destacando consistentemente    por un volumen mayor.
     * Esto podría alinearse con campañas promocionales o eventos específicos que impulsan las ventas en meses concretos.

```python
# Ventas Totales por Categoría
# Esto permite observar posibles outliers en las ventas por categoría.
plt.figure(figsize=(12, 6))
sns.boxplot(x='CATEGORY_DESCRIPTION', y='TOTAL_VALUE_SALES', data=reckitt_df)
plt.title('Boxplot de Ventas Totales por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Ventas Totales')
plt.xticks(rotation=45)
plt.show()
```

### <code><a>Insights<a/><code/> 
    * Las categorías líderes (Lavavajillas y AirCare) en ventas destacan claramente, y hay una brecha significativa entre las principales y las secundarias.
    * Esto resalta la necesidad de diversificar los ingresos, enfocándose en categorías menos explotadas con estrategias de marketing dirigidas.

```python
# Identificar Outliers Numéricamente
# Después de visualizar, se calcula los outliers usando la regla de 1.5 veces el rango intercuartil (IQR):

# Calcular el rango intercuartil
q1 = reckitt_df['TOTAL_VALUE_SALES'].quantile(0.25)
q3 = reckitt_df['TOTAL_VALUE_SALES'].quantile(0.75)
iqr = q3 - q1

# Límites de outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Filtrar los outliers
outliers = reckitt_df[(reckitt_df['TOTAL_VALUE_SALES'] < lower_bound) | 
                      (reckitt_df['TOTAL_VALUE_SALES'] > upper_bound)]

print(f"Número de posibles outliers: {len(outliers)}")
```
