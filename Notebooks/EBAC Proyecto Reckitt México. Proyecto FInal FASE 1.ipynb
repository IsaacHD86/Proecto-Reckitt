### * Estandarizacin de nombres de columnas:

```python
df_category.columns = df_category.columns.str.strip().str.upper()
df_sales.columns = df_sales.columns.str.strip().str.upper()
df_product.columns = df_product.columns.str.strip().str.upper()
df_segment.columns = df_segment.columns.str.strip().str.upper()
```

```python
# validacin de consistencia entre las columnas CATEGORY en df_product y ID_CATEGORY en df_category

# Categoras nicas en df_product
categ_product = df_product['CATEGORY'].unique()

# Categoras nicas en df_category
categ_category = df_category['ID_CATEGORY'].unique()

print('Categoras en df_product:', categ_product)
print('Categoras en df_category:', categ_category)
```

```python
# imputar CATEGORY en df_product basndote en FORMAT:
# Crear un diccionario con la relacin FORMAT -> ID_CATEGORY

format_category = {
    'BARRA': 1,
    'GEL': 2,
    'LIQUIDO': 3,
    'POLVO': 4,
    'TOALLAS': 5
}

# Imputar con .map() los valores de CATEGORY en df_product segn el valor de FORMAT
df_product['CATEGORY'] = df_product['FORMAT'].map(format_category)
df_product.head()
```

```python
# Verificar tipos de datos:

print(df_product['CATEGORY'].dtype)
print(df_category['ID_CATEGORY'].dtype)
```

```python
# Crear un diccionario con la relacin FORMAT -> CATEGORY de df_segment
format_category = {
    'BARRA': 1,
    'GEL': 2,
    'LIQUIDO': 3,
    'POLVO': 4,
    'TOALLAS': 5
}

# Imputar usando .map() con los valores de CATEGORY en df_segment segn el valor de FORMAT
df_segment['CATEGORY'] = df_segment['FORMAT'].map(format_category)

df_segment.head(10)
```

```python
# Verificar tipos de datos:

print(df_product['CATEGORY'].dtype)
print(df_category['ID_CATEGORY'].dtype)
print(df_segment['CATEGORY'].dtype)
```

# <li><code>Unir los DataFrames relevantes:<code>

```python
# Merge de df_product y df_segment:
df_product_segment = df_product.merge(df_segment, on=['CATEGORY', 'FORMAT', 'ATTR1', 'ATTR2', 'ATTR3'], how='left')
```

```python
# Merge con df_category:
df_merge = df_product_segment.merge(df_category, left_on='CATEGORY', right_on='ID_CATEGORY', how='left')
```

```python
df_merge.rename(columns={'ITEM': 'ITEM_CODE'}, inplace=True)
```

```python
#  Merge con df_sales:
df_merge = df_merge.merge(df_sales, on='ITEM_CODE', how='left')
```

```python
df_final = df_merge.merge(df_cal, on='WEEK', how='left')
```

```python
df_final.drop_duplicates(inplace=True)
```

```python
df_final.duplicated().sum()
```

```python
df_final.isnull().sum()
```

```python
# Imputacin de Valores nulos
most_frequent_week = df_final['WEEK'].mode()[0]
df_final['WEEK'].fillna(most_frequent_week, inplace=True)
```

```python
imp_null = df_final['TOTAL_UNIT_SALES'].median()
df_final['TOTAL_UNIT_SALES'].fillna(imp_null, inplace = True)
```

```python
imp_val = df_final['TOTAL_VALUE_SALES'].median()
df_final['TOTAL_VALUE_SALES'].fillna(imp_val, inplace = True)
```

```python
imp_unit = df_final['TOTAL_UNIT_AVG_WEEKLY_SALES'].median()
df_final['TOTAL_UNIT_AVG_WEEKLY_SALES'].fillna(imp_unit, inplace = True)
```

```python
imp_reg = df_final['REGION'].mode()[0]
df_final['REGION'].fillna(imp_reg, inplace = True)
```

```python
imp_year = df_final['YEAR'].mode()[0]
df_final['YEAR'].fillna(imp_year, inplace = True)
```

```python
imp_mon = df_final['MONTH'].mode()[0]
df_final['MONTH'].fillna(imp_mon, inplace = True)
```

```python
imp_wn = df_final['WEEK_NUMBER'].mode()[0]
df_final['WEEK_NUMBER'].fillna(imp_wn, inplace = True)
```

```python
imp_date = df_final['DATE'].mode()[0]
df_final['DATE'].fillna(imp_date, inplace = True)
```

```python
df_final.isnull().sum()
```

```python
df_final.info()
```

```python
# Eliminamos ID_CATEGORY porque contien los mismos valores que la columna CATEGORY
df_final.drop(columns=['ID_CATEGORY'], inplace = True)
```

```python
# Convertir columnas a tipos de datos ms eficientes
df_final['WEEK_NUMBER'] = df_final['WEEK_NUMBER'].astype('int64')
df_final['YEAR'] = df_final['YEAR'].astype('int64')
df_final['MONTH'] = df_final['MONTH'].astype('int64')
```

```python
df_final.rename(columns = {'CATEGORY_x': 'CATEGORY',
                           'CATEGORY_y': 'CATEGORY_DESCRIPTION'}, inplace = True)
```
